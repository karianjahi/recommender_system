```python
import pandas as pd
from sklearn.decomposition import NMF
import glob
```

```python
files = glob.glob("data/ml-latest-small/*csv")
```

```python
names = [file.split("/")[-1].split(".csv")[0] for file in files]
pd_dict = {}
for file, name in zip(files, names):
    pd_dict[name] = pd.read_csv(file)
```

```python
movies = pd_dict["movies"]
ratings = pd_dict["ratings"]
```

```python
ratings
```

```python
movies
```

```python
movies.set_index("movieId", inplace=True)
```

```python
ratings.set_index("movieId", inplace=True)
```

```python
ratings_with_movies = ratings.join(movies)
ratings_with_movies
```

```python
ratings_new = ratings_with_movies[["userId", "rating", "title"]]
ratings_new.index = range(len(ratings_new))
```

```python
ratings_reshaped = ratings_new.pivot_table(values="rating", columns="title", index="userId", aggfunc="mean")
ratings_reshaped
```

```python
ratings_reshaped.fillna(ratings_reshaped.mean(), inplace=True)
```

```python
ratings_reshaped
```

```python
nmf = NMF(n_components=20, max_iter=5000)
```

```python
nmf.fit(ratings_reshaped)
```

```python
Q = nmf.components_
Q
```

```python
Q_df = pd.DataFrame(Q).T
```

```python
Q_df.index = ratings_reshaped.columns
```
